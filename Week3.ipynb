```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Generate sample data
data = np.random.randn(100)
df = pd.DataFrame({'Values': data})

# Line Plot and Scatter Plot
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.figure(figsize=(12, 5))

# Line Plot
plt.subplot(1, 2, 1)
plt.plot(x, y, label='Sine Wave', color='b')
plt.title("Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()

# Scatter Plot
plt.subplot(1, 2, 2)
plt.scatter(x, np.cos(x), color='r', label='Cosine Points')
plt.title("Scatter Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()

plt.tight_layout()
plt.show()

# Histogram and Box Plot using Seaborn
plt.figure(figsize=(12, 5))

# Histogram
plt.subplot(1, 2, 1)
sns.histplot(df['Values'], bins=20, kde=True, color='g')
plt.title("Histogram")

# Box Plot
plt.subplot(1, 2, 2)
sns.boxplot(y=df['Values'], color='orange')
plt.title("Box Plot")

plt.tight_layout()
plt.show()
```
