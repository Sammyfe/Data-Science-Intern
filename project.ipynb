importing the necessary libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

Now we load the csv file

```python
# Read the CSV file
df = pd.read_csv('company_sales_data.csv')
```

```python
# Exercise 1: Line Plot of Total Profit
plt.figure(figsize=(10, 5))
plt.plot(df['month_number'], df['total_profit'], marker='o')
plt.title('Total Profit Over Months')
plt.xlabel('Month')
plt.ylabel('Total Profit')
plt.xticks(range(1, 13))  # Assuming months are numbered 1 to 12
plt.show()
```

```python
# Exercise 2: Subplot for Bathing Soap and Facewash
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 10))

# Bathing Soap Sales
ax1.plot(df['month_number'], df['bathingsoap'], marker='o', color='blue')
ax1.set_title('Bathing Soap Sales Over Months')
ax1.set_xlabel('Month')
ax1.set_ylabel('Units Sold')
ax1.set_xticks(range(1, 13))

# Facewash Sales
ax2.plot(df['month_number'], df['facewash'], marker='o', color='green')
ax2.set_title('Facewash Sales Over Months')
ax2.set_xlabel('Month')
ax2.set_ylabel('Units Sold')
ax2.set_xticks(range(1, 13))

plt.tight_layout()
plt.show()
```
